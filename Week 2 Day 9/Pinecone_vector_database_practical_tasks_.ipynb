```python
PINECONE_API_KEY="key"
pc = Pinecone(api_key=PINECONE_API_KEY)
```

```python
index_name = "developer-quickstart-py"

if not pc.has_index(index_name):
    pc.create_index_for_model(
        name=index_name,
        cloud="aws",
        region="us-east-1",
        embed={
            "model":"llama-text-embed-v2",
            "field_map":{"text": "chunk_text"}
        }
    )

index = pc.Index(index_name)  

records = [
    { "_id": "rec1", "chunk_text": "Long distance running builds exceptional cardiovascular endurance and strengthens the heart muscle with every training session.", "category": "running" },
    { "_id": "rec2", "chunk_text": "Cyclists improve their aerobic capacity and leg power by training at different intensity zones throughout the week.", "category": "cycling" },
    { "_id": "rec3", "chunk_text": "Swimming freestyle at high intensity burns more calories per hour than most other aerobic exercises available.", "category": "swimming" },
    { "_id": "rec4", "chunk_text": "Deadlifts and squats are compound gym exercises that activate multiple muscle groups simultaneously for maximum strength gains.", "category": "gym" },
    { "_id": "rec5", "chunk_text": "Proper running form includes a slight forward lean, relaxed shoulders, and landing with your foot under your hips.", "category": "running" },
    { "_id": "rec6", "chunk_text": "Mountain biking on rough terrain challenges both physical endurance and mental focus under constantly changing trail conditions.", "category": "cycling" },
    { "_id": "rec7", "chunk_text": "Butterfly stroke is the most technically demanding swimming style, requiring perfect timing between arm pulls and dolphin kicks.", "category": "swimming" },
    { "_id": "rec8", "chunk_text": "Progressive overload in gym training means gradually increasing weight or reps to continuously challenge your muscles and force growth.", "category": "gym" },
    { "_id": "rec9", "chunk_text": "Trail running on uneven surfaces improves ankle stability, balance, and engages far more muscles than running on flat roads.", "category": "running" },
    { "_id": "rec10", "chunk_text": "Competitive swimmers shave their bodies and wear specially designed suits to reduce drag and gain fractions of a second.", "category": "swimming" },
]

index.upsert_records("ns1", records)
```

```python
query = "Tell me about swimming techniques."

results = index.search(
    namespace="ns1",
    query={
        "top_k": 3,
        "inputs": {
            'text': query
        }
    }
)
```

Print search results.

```python
# place your code here
print(results)
```
